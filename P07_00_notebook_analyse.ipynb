# Analyse des tweets du jeu de données

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Réglage des graphiques
plt.style.use('seaborn-whitegrid')

plt.rc('font', size=12)
plt.rc('axes', titlesize=20)
plt.rc('axes', labelsize=20)
plt.rc('xtick', labelsize=12)
plt.rc('ytick', labelsize=12)
plt.rc('legend', fontsize=12)

dims_fig = (25,20)
```

```python
# Charger les données depuis le fichier csv
data = pd.read_csv('data/training.1600000.processed.noemoticon.csv', sep=',', encoding = "ISO-8859-1", header=None, names=['target', 'id', 'date', 'flag', 'user', 'text'])

data.head()
```

```python
data.info()
```

```python
data["target"].value_counts()
```

Il y a 800 000 tweets pour chaque target, 0 correspondant à un tweet négatif et 4 un tweet positif

```python
plt.figure(figsize=(20,10))
plt.hist(data["text"].apply(lambda x: len(x)).values,bins=200)
plt.title("Nombre de caractères des tweets")
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.hist(data["text"].apply(lambda x: len(x.split(" "))).values,bins=50)
plt.title("Nombre de mots des tweets")
plt.show()
```
